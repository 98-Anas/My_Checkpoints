```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("titanic-passengers.csv", sep=';')
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Sex'] = encoder.fit_transform(df['Sex'])
df['Survived'] = encoder.fit_transform(df['Survived'])
df
```

```python
df['Age'].fillna(df['Age'].mean(),inplace= True)
df_final = df.dropna(axis=1, how= 'any')
df_final
```

```python
plt.title('Passengers Survival')
survived_ppl = df_final['Survived'].value_counts()
survived_ppl.plot.bar(rot= 330)
print('0 : Survived | 1 : non-Survived')
```

```python
plt.title("Passenger's Class Survival Representation")
p_class = df_final['Pclass'].value_counts()
p_class.plot.bar(rot= 330)
print('1 : First Class | 2 : Second Class | 3 : Third Class')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df_final)
```

```python
def Split(word):
    return [char for char in word]

title = []
for i in range(len(df_final['Name'])):
    L1 = df_final['Name'][i].split()
    for elem in L1 :
        if elem == 'L.':
            continue
        for z in elem :
            if z == '.':
                title.append(elem)
    

title_dic = {'Title1': title}
t_df = pd.DataFrame(title_dic)
new_df = pd.concat([t_df,df_final],axis=1)
new_df
```

```python
Last_List = []
for i in range(len(new_df['Title1'])):
    Last_List.append(new_df['Title1'][i].split('.')[0])
Last_dic = {'Title': Last_List}
L_df = pd.DataFrame(Last_dic)
Last_df = pd.concat([L_df,new_df],axis=1)
Last_df = Last_df.drop('Title1', axis=1)
Last_df
```

```python
# Correlation between Title and Sex in relation to Survival of passenger
Corr = sns.FacetGrid(Last_df, col="Title", row="Sex")
Corr.map(sns.histplot, "Survived")
print('Survival = 0 : Survived | 1 : non-Survived')
print('     Sex = 0 : Female   | 1 : Male')
```
